```python
import numpy as np
import cPickle
from collections import defaultdict
import sys, re
import pandas as pd
import h5py
from random import shuffle
from scipy.spatial import distance
from matplotlib.mlab import PCA
import matplotlib.pyplot as plt
from numpy.linalg import norm
%matplotlib inline
```

# PREPROCESSING

```python
def clean_str(string, TREC=False):
    """
    Tokenization/string cleaning for all datasets except for SST.
    Every dataset is lower cased except for TREC
    """
    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)     
    string = re.sub(r"\'s", " \'s", string) 
    string = re.sub(r"\'ve", " \'ve", string) 
    string = re.sub(r"n\'t", " n\'t", string) 
    string = re.sub(r"\'re", " \'re", string) 
    string = re.sub(r"\'d", " \'d", string) 
    string = re.sub(r"\'ll", " \'ll", string) 
    string = re.sub(r",", " , ", string) 
    string = re.sub(r"!", " ! ", string) 
    string = re.sub(r"\(", " \( ", string) 
    string = re.sub(r"\)", " \) ", string) 
    string = re.sub(r"\?", " \? ", string) 
    string = re.sub(r"\s{2,}", " ", string)    
    return string.strip() if TREC else string.strip().lower()

def clean_str_sst(string):
    """
    Tokenization/string cleaning for the SST dataset
    """
    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)   
    string = re.sub(r"\s{2,}", " ", string)    
    return string.strip().lower()
```

```python
data_folder = ["rt-polarity.pos","rt-polarity.neg"] 
cv=10
clean_string=True
```

```python
revs = []
pos_file = data_folder[0]
neg_file = data_folder[1]
vocab = defaultdict(float)
```

```python
with open(pos_file, "rb") as f:
    for line in f:       
        rev = []
        rev.append(line.strip())
        if clean_string:
            orig_rev = clean_str(" ".join(rev))
        else:
            orig_rev = " ".join(rev).lower()
        words = set(orig_rev.split())
        for word in words:
            vocab[word] += 1
        datum  = {"y":1, 
                  "text": orig_rev,                             
                  "num_words": len(orig_rev.split()),
                  "split": np.random.randint(0,cv)}
        revs.append(datum)
```

```python
with open(neg_file, "rb") as f:
    for line in f:       
        rev = []
        rev.append(line.strip())
        if clean_string:
            orig_rev = clean_str(" ".join(rev))
        else:
            orig_rev = " ".join(rev).lower()
        words = set(orig_rev.split())
        for word in words:
            vocab[word] += 1
        datum  = {"y":0, 
                  "text": orig_rev,                             
                  "num_words": len(orig_rev.split()),
                  "split": np.random.randint(0,cv)}
        revs.append(datum)
```

```python
shuffle(revs)
```

```python
max_l = np.max(pd.DataFrame(revs)["num_words"])
```

```python
def get_W(word_vecs, k=300):
    """
    Get word matrix. W[i] is the vector for word indexed by i
    """
    vocab_size = len(word_vecs)
    word_idx_map = dict()
    W = np.zeros(shape=(vocab_size+1, k))            
    W[0] = np.zeros(k)
    i = 1
    for word in word_vecs:
        W[i] = word_vecs[word]
        word_idx_map[word] = i
        i += 1
    return W, word_idx_map
```

```python
k = max_l
word_vecs = vocab
vocab_size = len(word_vecs)
word_idx_map = dict()
W = np.zeros(shape=(vocab_size+1, k))    
W[0] = np.zeros(k)
i = 1
for word in word_vecs:
    W[i] = word_vecs[word]
    word_idx_map[word] = i
    i += 1
```

```python
train = []
target = []
test = []
test_target = []

for rev in revs:
    if rev["split"] == 1: 
        test.append([1] * (max_l+5))
        test_target.append(rev["y"]+1)
        for i, word in enumerate(rev['text'].split()):
            test[-1][i+5] = word_idx_map[word] + 1

    else:
        train.append([1] * (max_l+5))
        target.append(rev["y"]+1)
        for i, word in enumerate(rev['text'].split()):
            train[-1][i+5] = word_idx_map[word] + 1
```

# WRITE DATA

```python
f = h5py.File('movies.hdf5', 'w')
```

```python
f['train'] = train
f['target'] = target
f['test'] = test
f['test_target'] = test_target
```

```python
f.close()
```

# FINDING COOL WORDS

```python
f = h5py.File('movieweights.h5', 'r')
lookup = f["lookup"]
```

```python
def findSimilarWordsDotProduct(word,n = 10):
    arr = np.dot(np.array(lookup),np.array(lookup[word_idx_map[word]]))
    indices = arr.argsort()[-n:][::-1]
    return([vocab.keys()[i-1] for i in indices])
```

```python
def findSimilarWordsCosineDistance(word,n = 10):
    dists = [0 for x in range(len(lookup))]
    for x in range(len(lookup)):
        dists[x] = distance.cosine(lookup[word_idx_map[word]],lookup[x])
    indices = np.array(dists).argsort()[:11]
    return([vocab.keys()[i-1] for i in indices])
```

```python
def plotNeighbor(word,radius = 1):
    targetx = results.Y[word_idx_map[word]][0]
    targety = results.Y[word_idx_map[word]][1]
    x = []
    y = []
    for i in range(1,len(results.Y)):
        xpoint = results.Y[i][0]
        ypoint = results.Y[i][1]
        if norm(np.array([xpoint,ypoint])-np.array([targetx,targety])) < radius:
            x.append(xpoint)
            y.append(ypoint)
    plt.scatter(x,y)
    for i in range(1,len(results.Y)):
        xpoint = results.Y[i][0]
        ypoint = results.Y[i][1]
        if norm(np.array([xpoint,ypoint])-np.array([targetx,targety])) < radius:
            plt.annotate(vocab.keys()[i-1],xy = (xpoint,ypoint), textcoords = 'offset points')
```
